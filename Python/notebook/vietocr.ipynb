```python
from vietocr.tool.predictor import Predictor
from vietocr.tool.config import Cfg
```

```python
config_path = r"C:\Users\caotr\Downloads\AI\modelconfig\config.yml"
weight_path = r"C:\Users\caotr\Downloads\AI\modelconfig\transformerocr.pth"
model_config = Cfg.load_config_from_file(config_path)
model_config['weights'] = weight_path
```

```python
model_config['device'] = 'cuda:0'
model_config['cnn']['pretrained'] = False
model_config['predictor']['beamsearch'] = True
```

```python
detector = Predictor(model_config)
```

```python
img = r"C:\Users\caotr\OneDrive\Hình ảnh\test1.png"
img = Image.open(img)
plt.imshow(img)
s = detector.predict(img)
s
```
