```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gradient_descent(x, y):
    m = b = 0
    it = 1000
    n = len(x)
    learning_rate = 0.08
    plt.scatter(x, y)
    for i in range(it):
        y_predict = m * x + b
        cost = (1/n) * sum([val**2 for val in (y - y_predict) ])
        md = -(2/n) * sum(x * (y - y_predict))
        bd = -(2/n) * sum(y - y_predict)
        m = m - learning_rate * md
        b = b - learning_rate * bd
        plt.plot(x, y_predict, color='green', alpha=0.1)

    print("cost = ", cost)
    print("y = ", m, "x + ", b)
    

        
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
```

```python
gradient_descent(x,y)
```

```python
import math
from sklearn import linear_model
def gradient_descent2(x, y):
    m = b = 0
    it = 400000
    n = len(x)
    learning_rate = 0.0002
    plt.scatter(x, y)
    r = linear_model.LinearRegression()
    r.fit(x.reshape(-1,1), y)
    plt.plot(x, r.predict(x.reshape(-1,1)), color='red')
    print("y = ", r.coef_, "x + ", r.intercept_)
    cost_previous = 0
    for i in range(it):
        y_predict = m * x + b
        cost = (1/n) * sum([val**2 for val in (y - y_predict) ])
        if math.isclose(cost, cost_previous, rel_tol=1e-20):
            break
        cost_previous = cost
        md = -(2/n) * sum(x * (y - y_predict))
        bd = -(2/n) * sum(y - y_predict)
        m = m - learning_rate * md
        b = b - learning_rate * bd
        if i % 1000 == 0:
            plt.plot(x, y_predict, color='green', alpha=0.1)
    plt.plot(x, y_predict, color='green')
    print("cost = ", cost)
    print("y = ", m, "x + ", b)

gradient_descent2(x,y)
```
