```python
import pandas as pd
import csv
import numpy as np
import os
```

```python
img = r"meme dataset 1"
csv_list = [
    r"database/webscrapping1.csv",
    r"database/webscrapping2.csv",
    r"database/webscrapping3.csv",
    r"database/webscrapping4.csv",
]

# append all csv
df = pd.concat([pd.read_csv(f) for f in csv_list], ignore_index=True)


# Drop duplicate
df = df.drop_duplicates(subset=["post_url"])

# Drop null
df = df.dropna(subset=["post_url"])
```

```python
df
```

```python
with open("database/webscrapping_final.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["ID", "post_url", "image_url", "post_date", "post_content", "post_likes", "post_comments"])
    for i, row in df.iterrows():        
        img_id = row["post_url"].split("/")[6]
        # check if image id is in meme dataset
        exist_os_path1 = os.path.exists(os.path.join(img, img_id + "_lowquality.jpg"))
        exist_os_path2 = os.path.exists(os.path.join(img, img_id + ".jpg"))
        if exist_os_path1 or exist_os_path2:
            writer.writerow([img_id, row["post_url"], row["post_image_lowquality"], row["post_date"], row["post_content"], row["post_likes"], row["post_comments"]])
            # change image name
            if exist_os_path1:
                os.rename(os.path.join(img, img_id + "_lowquality.jpg"), os.path.join(img, img_id + ".jpg"))
        else:
            continue
```
