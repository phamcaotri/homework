```python
accept_header = np.array(['Sơ lược', 'Nội dung', 'Công dụng', 'Xuất hiện', 'ý', 'Lỗi', 'Cách', 'Trang bị'])
file_name = 'doraemon_gadgets_vietnamese_dataset.csv'
mainpages_url = ['https://doraemon.fandom.com/vi/wiki/Th%E1%BB%83_lo%E1%BA%A1i:B%E1%BA%A3o_b%E1%BB%91i','https://doraemon.fandom.com/vi/wiki/Th%E1%BB%83_lo%E1%BA%A1i:B%E1%BA%A3o_b%E1%BB%91i?from=Si%C3%AAu+tr%E1%BB%99m+Deluxe']
with open(file_name, 'w', newline='', encoding='utf-16') as csvfile:
    for mainpage_url in mainpages_url:
        mainpage = requests.get(mainpage_url)
        mainpage_contents = BeautifulSoup(mainpage.content, 'html.parser')
        writer = csv.writer(csvfile, delimiter='\t')
        writer.writerow(['Tên', 'Sơ lược', 'Nội dung', 'Công dụng', 'Xuất hiện', 'Lưu ý', 'Lỗi', 'Cách sử dụng', 'Trang bị'])
        for mpc in mainpage_contents.find_all('a', class_='category-page__member-link'):
            if 'Tập tin' in mpc['title'] or 'Thể loại' in mpc['title']:
                continue
            # write ten
            print('mpc["title"]: ', mpc['title'])
            data = {'Tên': mpc['title']
                     , 'Sơ lược': ''
                     , 'Nội dung': ''
                     , 'Công dụng': ''
                     , 'Xuất hiện': ''
                     , 'Lưu ý': ''
                     , 'Lỗi': ''
                     , 'Cách sử dụng': ''
                     , 'Trang bị': ''}
            contents = BeautifulSoup(requests.get('https://doraemon.fandom.com' + mpc['href']).content, 'html.parser')
            for header in contents.find_all('h2'): 
                if 'Sơ lược' in header.text:
                    data['Sơ lược'] = header.find_next_sibling('p').text if header.find_next_sibling('p') else ''
                elif 'Nội dung' in header.text:
                    data['Nội dung'] = header.find_next_sibling('p').text if header.find_next_sibling('p') else ''
                elif 'Công dụng' in header.text:
                    data['Công dụng'] = header.find_next_sibling('p').text if header.find_next_sibling('p') else ''
                elif 'Xuất hiện' in header.text:
                    data['Xuất hiện'] = header.find_next_sibling('p').text if header.find_next_sibling('p') else ''
                elif 'ý' in header.text:
                    data['Lưu ý'] = header.find_next_sibling('p').text if header.find_next_sibling('p') else ''
                elif 'Lỗi' in header.text:
                    data['Lỗi'] = header.find_next_sibling('p').text if header.find_next_sibling('p') else ''
                elif 'Cách' in header.text:
                    data['Cách sử dụng'] = header.find_next_sibling('p').text if header.find_next_sibling('p') else ''
                elif 'Trang bị' in header.text:
                    data['Trang bị'] = header.find_next_sibling('p').text if header.find_next_sibling('p') else ''
                else:
                    continue

            writer.writerow([data['Tên'], data['Sơ lược'], data['Nội dung'], data['Công dụng'], data['Xuất hiện'], data['Lưu ý'], data['Lỗi'], data['Cách sử dụng'], data['Trang bị']])
        


```
