```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('database/diemtbhk1_2023uit.csv')
df
```

```python
df[df['HoTen'].str.contains('Minh Trí') & df['Lop'].str.contains('CNTT2022.2')]
```

```python
count_df = pd.DataFrame(columns=['DiemTB', 'SoLuong'])
count_df = df.groupby('DiemTB').size().reset_index(name='SoLuong')
count_df
```

```python
count_df.describe()
```

```python
x = count_df['DiemTB']
y = count_df['SoLuong']
fig, ax = plt.subplots(figsize=(20, 15))
plt.xlabel('Điểm')
plt.ylabel('Số lượng')
plt.title('Phân tích điểm trung bình học kỳ 1 năm 2023 UIT')
plt.xticks(np.arange(0, 10 + 0.25, 0.25))
plt.bar(x, y, width = 0.01)
plt.show()
```
