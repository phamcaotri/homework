```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("database/canada_per_capita_income.csv")
df.head(5)
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df.get('year'), df.get('per capita income (US$)'), color='red', marker='.')
r = linear_model.LinearRegression()
r.fit(df[['year']], df[['per capita income (US$)']])
plt.plot(df.get('year'), r.predict(df[['year']]), color='blue')
```
