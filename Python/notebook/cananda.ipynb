```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("database/canada_per_capita_income.csv")
df.head(5)
```

```python
x = df[['year']]
y = df[['per capita income (US$)']]
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(x, y, color='red', marker='.')
r = linear_model.LinearRegression()
r.fit(x.values, y.values)
plt.plot(x, r.predict(x), color='blue')
r.score(x.values, y.values)
```

```python
print(r.predict(X=[[2020]]).item())
```

```python
import pickle
with open('models/canada_linear_regression_model.pickle', 'wb') as f:
    pickle.dump(r, f)
with open('models/canada_linear_regression_model.pickle', 'rb') as f:
    model = pickle.load(f)

print(model.predict(X=[[2020]]).item())
```
