```python
file_name = 'database/webscrapping03082023.csv'
with open(file_name, 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['post_url', 'post_content', 'post_date', 'post_image', 'post_image_lowquality', 'post_video', 'post_likes', 'post_comments', 'post_shares', 'comments']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    for i, post in enumerate(posts):
        print(f'Writing post {i+1}')
        # Check if post has image or not
        if post['image'] == None:
            post['image'] = 'No image'

        # Check if post has image with low quality or not
        if post['image_lowquality'] == None:
            post['image_lowquality'] = 'No image'

        # Check if post has video or not
        if post['video'] == None:
            post['video'] = 'No video'

        # Get comments, save them as a dictionary
        """
        format:
        {
            'commenter_name': {
                'comment': 'comment_content',
                'replies': {
                    'replier_name': {
                        'reply': 'reply_content'
                    }
                }
            }
        }
        """

        comment_tree = {}
        for comment in post['comments_full']:
            comment_tree[comment['commenter_name']] = {
                'comment': comment['comment_text'].replace('\n', ' '),
                'replies': {}
            }
            for reply in comment['replies']:
                comment_tree[comment['commenter_name']]['replies'][reply['commenter_name']] = {
                    'reply': reply['comment_text'].replace('\n', ' ')
                }
        
        # download images
        image_name = post['post_id'] + '.jpg'
        image_lowquality_name = post['post_id'] + '_lowquality.jpg'
        if post['image'] != 'No image':
            response = requests.get(post['image'])
            with open('images/' + image_name, 'wb') as f:
                f.write(response.content)
        if post['image_lowquality'] != 'No image':
            response_lowquality = requests.get(post['image_lowquality'])
            with open('images/' + image_lowquality_name, 'wb') as f:
                f.write(response_lowquality.content)
        if post['video'] != 'No video':
            response_video = requests.get(post['video'])
            with open('images/' + post['post_id'] + '.mp4', 'wb') as f:
                f.write(response_video.content)
        data = {
            'post_url': post['post_url'],
            'post_content': post['text'].replace('\n', ' '),
            'post_date': post['time'].strftime('%Y-%m-%d %H:%M:%S'),
            'post_image': post['image'],
            'post_image_lowquality': post['image_lowquality'],
            'post_video': post['video'],
            'post_likes': post['likes'],
            'post_comments': post['comments'],
            'post_shares': post['shares'],
            'comments': comment_tree
        }

        writer.writerow(data)
```
